The Museum of Motoring History is preparing for an exhibition about vehicles from the 1970s. The curator wants to create an interactive exhibit for visitors to learn about the fuel efficiency of many popular cars. They have a data set of car fuel efficiency from 1970 to 1982 in which the fuel efficiency is measured in miles per gallon (MPG). A higher MPG means the car is more efficient. What visualisations could the exhibit include to engage visitors? This is where pointplots can be used as a replacement for bar charts.

### 1. Prepare your workstation

```python
# Import Matplotlib, Seaborn, and Pandas.
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read the CSV file.
mpg = pd.read_csv('mpg.csv')

# View the DataFrame.
print(mpg.shape)
print(mpg.dtypes)
print(mpg.columns)
mpg.head()
```

### 2. Create a lineplot

A lineplot displays data with a straight or curved line by connecting individual lines. Lineplots are mainly used to track changes over periods of time. However, a lineplot is not a time series. 

In preparation for the exhibition, create a lineplot to compare the fuel efficiency of popular car models between 1970 and 1982. The code snippet can be written as follows:

Specify the alias of the Seaborn library.
Add the lineplot() function.
Specify the parameters as x='model_year', y='mpg', and data=mpg.

```python
# Create a lineplot for fuel efficiency.
sns.lineplot(x='model_year', y='mpg', data=mpg)
```

The figure displays the model_year on the x-axis and the fuel efficiency (mpg) on the y-axis. However, you cannot interpolate along the line between the years 1972 and 1973 since there isnâ€™t such a thing as a â€˜1972.5â€™ car model. Therefore, the lines create a correlation (effect) that is misleading. 

### 3. Create a barplot

# Barplots
A barplot is commonly used to indicate a relationship between numeric and categoric variables. For example, you might use a barplot to visualise the number of students in each grade based on gender. Different hues (colours) can be used to distinguish between grades and gender. The number of students will be on the y-axis, while the grades and gender will be displayed on the x-axis.

After looking at the output, decide whether a barplot will display the information better than a lineplot when comparing the fuel efficiency of popular car models between 1970 and 1982. The code snippet can be written as follows:

Specify the alias of the Seaborn library.
Add the barplot() function.
Specify the parameters as x='model_year', y='mpg', and data=mpg.

```python
# Create a barplot for fuel efficiency.
sns.barplot(x='model_year', y='mpg', data=mpg)
```

Again, the figure displays the model_year on the x-axis and the fuel efficiency (mpg) on the y-axis. Each year is represented by a different colour, creating a very â€˜busyâ€™ visualisation. Also, the trend is perhaps not immediately obvious and might be tricky for an audience to read. The error bars do not indicate a significant difference between the bars.

### 4. Create a pointplot

Pointplots
Create a pointplot to compare the fuel efficiency of popular car models between 1970 and 1982. The code snippet can be written as follows:

Specify the alias of the Seaborn library.
Add the pointplot() function.
Specify the parameters as x='model_year', y='mpg', and data=mpg.

```python
# Create a pointplot for fuel efficiency.
sns.pointplot(x='model_year', y='mpg', data=mpg, color='red')
```

What if we want to display the fuel efficiency of popular car models per manufacturing country? The code snippet can be written as follows:

Specify the alias of the Seaborn library.
Add the pointplot() function.
Specify the parameters as x='model_year', y='mpg', hue='origin', and data=mpg.

```python
# Create a pointplot for fuel efficiency by manufacturing country.
sns.pointplot(x='model_year', y='mpg', hue='origin', data=mpg)
```

The USA was behind Japan and Europe in terms of fuel efficiency. This could be a good starting point to further examine the data and see what the reason for this was (e.g. heavier vehicles, better performance, unit conversion errors,)
