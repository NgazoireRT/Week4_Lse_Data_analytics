```python
# Add Labels, ticks and tick labels 

# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# create and empty plot

fig, ax= plt.subplots()

# create variables. 

x = np.linspace(0, 10, 101)
y = np.exp(x)

# create plot 
ax.plot(x, y)
```

Since the chart is missing labels, the data makes very little sense. Are we looking at sales growth over the last 10 months or the exponential growth of bacteria? Imagine you encounter a chart like this in a publication. What additional information or context would you need to be able to make sense of the visualisation? 

Let’s assume that it is showing the growth rate of bacteria, where:

- the y-axis is the number of bacteria
- the x-axis is the time elapsed, in minutes.

# Add labels, ticks, and tick labels

```python
# Set labels and title.
ax.set_xlabel("Time Elapsed (minutes)")
ax.set_ylabel("Count")
ax.set_title("Bacteria Growth")

# View the plot.
fig
```

This plot has major tick marks on the x-axis in multiples of two and the y-axis has them in multiples of 5,000. These major tick marks are automatically labelled.

To set the position of tick marks on an axis, use the set_xticks() or set_yticks() method. Both of these methods take an array of floats (decimal numbers) as the first positional argument and optionally the keyword argument minor. In other words, it is optional to include minor ticks on the plot (minor=True or minor=False). The array argument sets the position of the tick marks along that axis. Minor is False by default and should be set to True to set minor tick marks instead of major.

To continue with the bacteria example, let’s reduce the number of major tick marks on the x-axis to just three, at positions 0, 5, and 10. We’ll then add minor tick marks in between. Enter the following code:

Use the set_xticks() method on ax, and specify:
- xticks(0, 5, 10): set ticks at 0, 5, and 10 on the x-axis
- xticks(1, 2, 3, 4, 6, 7, 8, 9): set ticks at 1 to 4 and 6 to 7 on the x axis. Add minor=True to include minor ticks at these points.

View the figure.

# Adjusting the ticks

```python
# add ticks to the x-axis

ax.set_xticks([0, 5, 10])
ax.set_xticks([1, 2, 3, 4, 5, 6, 7, 8, 9], minor=True)

#view the figure
fig
```

# Customising labels

```python
# Add ticks labels to x-axis.
ax.set_xticklabels(['0.0', '5.0', '10.0'])

# View the figure.
fig
```

```python
# Add ticks labels to x-axis.
ax.set_xticklabels(['0', '1', '2', '3', '4', '6', '7', '8', '9'], minor=True)

# View the figure.
fig
```

# Generate legends 

```python
# Create new variables.
y_no_food = np.exp(x)
y_food = np.exp(x + 1.01)

# Create an empty plot.
fig, ax = plt.subplots()

# Plot the new variables and add the label to the legend.
lines = ax.plot(x, y_no_food)
line = lines[0]

# Set the line variable.
line.set_label("No Food")

# Add legend.
ax.legend()
```

# Adjusting the legends 

Next, we can plot the other line of the bacterial growth for when they’re given food. We’ll look at another method of labelling a line: passing it as the label keyword argument to the plot method. This combines the plotting and labelling into one step. The code snippet can be written as follows:

- Plot the second line to represent the y variable of With Food.
- Create a new legend and specify the position of the legend with the loc='center left' parameter.
- View the figure.

```python
# Plot second line.
ax.plot(x, y_food, label="With Food")

# Create legend.
ax.legend()

# Specify legend position.
ax.legend(loc='center left')

# View figure.
fig
```
