# 4.1.2 - Matplotlib Foundations

```python
# Import Matplotlib as plt.
import matplotlib.pyplot as plt
```

# One positional argument

```python
# Create an empty figure with two axes.
fig, ax = plt.subplots()
```

```python
# Specify only one argument = y-values.
ax.plot([0, 10, 20, 30, 40])

# View the figure.
fig
```

Note that the y-axis has the values 0 to 40, while the x-axis has the values 0 to 4.0. We only specified the y-values and therefore, Python created default values for the x-axis.

# Two positional arguments

```python
# Specify two arguments = x-values and y-values

ax.clear()
ax.plot([0, 1, 2, 3, 4], [4, 3, 2, 1, 0])

# view the figure
fig
```

You can see on the plot when x=0 y=4 and so forth. Let’s take it one step further and investigate how to change the appearance of the figure.

#  three positional argument

```python
# Specify three arguments = x-values, y-values, and marker.
ax.clear()
ax.plot([0, 1, 2, 3, 4], [4, 3, 2, 1, 0], 'og-')

# View the figure.
fig
```

If you wanted to change the format to blue cross markers without a line between markers, the code snippet can be written as follows:

Clear the axis to indicate that you want to start with a clean slate (i.e. a figure with no axes) with the ax.clear() method.
Specify the values for the x-axis as 0, 1, 2, 3, 4 and the y-axis as 4, 3, 2, 1, 0 with the ax.plot() method. After the values, add the following string:
x: cross markers
b: blue colour
Remember the quotation marks for the string.
View the figure.

```python
# Specify three arguments (x-values, y-values, and marker) without a line.
ax.clear()
ax.plot([0, 1, 2, 3, 4], [4, 3, 2, 1, 0], 'xb')

# View the figure.
fig
```

# Named Arguments 

```python
# Specify various named arguments.
ax.clear()
ax.plot([0, 1, 2, 3, 4],
        [4, 3, 2, 1, 0],
        color='#94CEE4',
        marker='x',
        markeredgecolor='red',
        drawstyle='steps',
        linestyle=':')

# View the figure.
fig
```

```python
# Create a pairplot 

sns.pairplot(penguins)
```

A figure in the form of a map is returned, indicating all the numeric variables in the data set mapped against each other. You can see the patterns, anomalies, and outliers in the data set at a glance. The map can be read by cross-referencing the y-axis and x-axis as with a normal scatterplot. For example, there is a strong positive correlation between flipper_length_mm (x-axis) and body_mass_g (y-axis).

```python
# Create a pairplot with KDE (kernel density estimate).

sns.pairplot(penguins, hue='species')
```

This returns an even more granular visualisation of the data on a single plane. Here, we have used species to highlight the differences. Each species is indicated by a different colour. For example, there is a strong positive correlation between flipper_length_mm (x-axis) and body_mass_g (y-axis) for all three species.

```python
# Create a pairplot with the facetgrid method.

sns.pairplot(penguins, hue='species', diag_kind='hist')
```

Here we can see that the marginal plots along the diagonal are displayed as histograms with the hue differentiated by species.

```python
# Create a pairplot and indicate the height.

sns.pairplot(penguins, hue='species', diag_kind='hist', height=4)
```

Adjusting the size may be necessary when the space allocated for the final visualisation is restricted (for example, if the visualisation is to be published as part of an article or report). Play around with the size to see the difference.

# 4.2.3 - Customising Styles and Colours, titles and labels

# Changing plot style and colours with Seaborn

For better visibility, let’s adjust the figure size and plot our first scatterplot using the default darkgrid theme. The code snippet can be written as follows:

- Set the figure size with the set() method to 5 × 5. You can play around with the size to understand the impact on clarity.
- Set the tick style and the colour to darkgrid with the set_style function.
- Plot the scatterplot by specifying the data as penguins, the x as bill_length_mm, y as flipper_length_mm, and hue as species.
- Execute the code snippet.

```python
# Set the figure size.
sns.set(rc={'figure.figsize':(5, 5)})

# Set the tick style.
sns.set_style('ticks')

# Set the colour style.
sns.set_style('darkgrid')

# Plot the scatterplot.
sns.scatterplot(data=penguins, x='bill_length_mm', 
                y='flipper_length_mm', hue='species')
```

```python
# Customise individual attributes

# Set the style of ticks.
sns.set_style('ticks', {'axes.facecolor': '#dddddd',
                        'axes.spines.top': False,
                        'axes.spines.right': False})

# Create the scatterplot.
sns.scatterplot(data=penguins, x='bill_length_mm',
                y='flipper_length_mm', hue='species')
```

```python
# Change the colour of the data series in Seaborn

# Set the style first.
sns.set_style('ticks')

# Specify the Pastel palette.
sns.scatterplot(data=penguins, x='bill_length_mm',
                y='flipper_length_mm', hue='species',
                palette='muted')
```

```python
# Customise labels in Seaborn

# Plot the scatterplot.
ax = sns.scatterplot(data=penguins, x='bill_length_mm',
                     y='flipper_length_mm', hue='species',
                     palette='colorblind')

# Specify the labels.
ax.set_xlabel("Bill length (mm)")
ax.set_ylabel("Flipper length (mm)")
ax.set_title("Penguins Bill vs. Flipper Length")
```

# 4.2.6 Placing annotations demonstration

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Plot a simple line. 

fig, ax = plt.subplots()
ax.plot([4, 3, 2, 1, 0])
```

```python
# Set the annotation origin

data_ann =ax.annotate('Data 2, 2', xy=(2,2))

fig
```

```python
# Axes pixels cordinate system
pixel_ann = ax.annotate('Pixels 2, 2', xy=(2,2), xycoords='axes pixels')

fig
```

```python
# Fixing the test alignment

data_ann.remove()
pixel_ann.remove()
centered_ann = ax.annotate("Centered Data 2, 2",
                           xy=(2,2), horizontalalignment='center',
                           verticalalignment='center')
fig
```

# adding arrows to your plot

Add arrows to your plot to point to specific features. To add an arrow, the keyword argument arrowprops is passed to the annotate method, which is a dictionary that contains properties used to define how the arrow is displayed. At a minimum, you can use these properties:

- facecolor: the colour of the arrow.
- shrink: the percentage to ‘pull’ the arrow in from the text and where it points. (A good value is 0.05, or 5%.)
- Provide separate positions for the point of the arrow and the text position. (The arrow will be drawn between these two points.)
- The data point (and tip of the arrow) is positioned with the xy keyword argument, while the text is positioned with the xytext keyword argument.
- We can set the data point coordinate system and text with the keyword arguments xycoords and textcoords respectively. The coordinate system is ‘data’ by default.
- Putting this all together, we can add an arrow annotation by calling the annotate method.

```python
centered_ann.remove()
ax.annotate("Look at point 1, 3", xy=(1, 3),
            xytext=(1, 1.5),
            arrowprops={'facecolor': 'black', 'shrink': 0.05},
            horizontalalignment='center', verticalalignment='top')

fig
```
