# Setting the resolution 
First, let's practise setting the DPI of the figure when itâ€™s created simply using the DPI argument in the subplots() function. Launch Jupyter Notebook and create a new Python3 file, and then create a six by four inch figure at 150 DPI. 

```python
# Import Seaborn and Matplotlib.
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(dpi=150)

fig.set_size_inches(6, 4)
```

```python
fig, ax = plt.subplots(dpi=100)

fig.set_size_inches(8, 6)
```

```python
# Saving the visualisation. 

fig.savefig('my_plot.png')

fig.savefig('my_plot.jpg')
```

```python
fig.savefig('my_plot_print.png', dpi=300)

fig.savefig('my_plot_web.png', dpi=72)
```
